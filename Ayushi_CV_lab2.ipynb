# **EXPERIMENT 2 -MORPHOLOGICAL OPERATIONS**

**CONCLUSION -Morphological operations offer powerful tools for image processing, enabling the extraction of essential features and patterns while filtering out noise and artifacts. By utilizing dilation, erosion, and gradient techniques, morphological operations enhance image quality and aid in object detection and segmentation tasks.**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('/content/good.jpg')
plt.imshow(img)
```

##**Dilation**

```python
kernel = np.ones((7,7),np.uint8)
img_dilation = cv2.dilate(img, kernel, iterations=1)
plt.imshow(img_dilation)
```

## **Erosion**

```python
kernel = np.ones((5,5),np.uint8)
img_erosion = cv2.erode(img_dilation, kernel, iterations=1)
plt.imshow(img_erosion)
```

## **LAB TASK 1**

```python
circle = cv2.imread('/content/images.jpg')
plt.imshow(circle)
```

```python
kernel = np.ones((10,10),np.uint8)
img_erosion = cv2.erode(circle, kernel, iterations=1)
plt.imshow(img_erosion)
```

```python
# Define the structuring element
kernel = cv2.getStructuringElement(cv2.MORPH_RECT,(16,16))
# Apply the opening operation
opening = cv2.morphologyEx(img_erosion, cv2.MORPH_OPEN, kernel)
plt.imshow(opening)
```

## **Lab Task 2**

```python
finger = cv2.imread('/content/fingerprint.png')
plt.imshow(finger)
```

```python
kernel = np.ones((2,2),np.uint8)
img_erosion = cv2.erode(finger, kernel, iterations=1)
plt.imshow(img_erosion)
```

```python
kernel = cv2.getStructuringElement(cv2.MORPH_RECT,(3,3))
opening = cv2.morphologyEx(img_erosion, cv2.MORPH_OPEN, kernel)
plt.imshow(opening)
```

```python
kernel = np.ones((2,2),np.uint8)
img_dilation = cv2.dilate(opening, kernel, iterations=1)
plt.imshow(img_dilation)
```

```python
closing = cv2.morphologyEx(opening, cv2.MORPH_CLOSE, kernel)
plt.imshow(closing)
```

```python
concatenated_image = cv2.hconcat([cv2.resize(finger,(200,200)),cv2.resize(opening,(200,200)),cv2.resize(closing,(200,200))])
plt.imshow(concatenated_image)
```

## **Lab Task 3**

```python
brain =cv2.imread("/content/brain.jpg.png")
plt.imshow(brain)
```

```python
# Perform grayscale dilation
dilation_kernel = np.ones((3, 3), np.uint8)
dilation = cv2.dilate(brain, dilation_kernel, iterations=1)
```

```python
# Perform grayscale erosion
erosion_kernel = np.ones((3, 3), np.uint8)
erosion = cv2.erode(brain, erosion_kernel, iterations=1)
```

```python
# Calculate morphological gradient
gradient = cv2.morphologyEx(brain, cv2.MORPH_GRADIENT, np.ones((3, 3), np.uint8))
```

```python
plt.figure(figsize=(15, 12))

plt.subplot(1, 4, 1)
plt.imshow(brain, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 4, 2)
plt.imshow(dilation, cmap='gray')
plt.title('Grayscale Dilation')

plt.subplot(1, 4, 3)
plt.imshow(erosion, cmap='gray')
plt.title('Grayscale Erosion')

plt.subplot(1, 4, 4)
plt.imshow(gradient, cmap='gray')
plt.title('Morphological Gradient')
```
